```python
import requests

try:
    url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"
    resp = requests.get(url)
    print(f"Response code: {resp.status_code}")
    if resp.status_code == 200:
        data = resp.json()
        print
        print(f"data: {data[:5]}")
        print(f"Total no.of countries returned: {len(data)}")
    else:
        print(resp.text)

except Exception as e:
    print(f"Exception occured: {e}")


```

```python
import pandas as pd
countriesList = []
country_dict={}

for i in range(0,len(data)):
    row = data[i]
    nameObj=row.get("name")
    country_dict = {}
    country_dict['commonName'] = nameObj.get("common")
    country_dict['population']=row.get("population")
    country_dict['region']=row.get("region")
    country_dict['area']=row.get("area")

    countriesList.append(country_dict)

print(f"\n countriesList: {countriesList[:5]}")    

df = pd.DataFrame(countriesList)
print(df.head())
```
