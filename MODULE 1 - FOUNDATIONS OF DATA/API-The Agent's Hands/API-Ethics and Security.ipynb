### Real Life Scenario
**Problem** : You have added your API_KEY in a file and used it in your code. Now you have pushed everything in github. Next morning you got a credit card bill of 5000. Now how did this happen? This is due to, you pushing the api key into github.

**Solution**: These type of problems can be prevented by adding a .gitignore file to you github repo, and mentioning the .env file in it. This will tell the git to ignore .env files from committing.

## How to prevent the above problems?
1. Do not share the API keys to anyone
2. Do any have any screenshots or images or api keys, do not screen share with api keys on your screen.
3. Do not push .env files to github.
4. If you have somehow exposed you API key, them immediately delete the key.
5. What if you have exposed your API key and you are unaware of it? In such scenarios, it is always advised to do a key rotation. **Key rotation** means, deleting and creating new keys for every 30 days or 15 days.
6. Avoid handling everything with a single API key. Like while creation, we can give less privilages to the API key that we create. This approach is known as **Least privilage Principle**.Like creating an API key with read-on;y acecess. This will limit the exploitation, by the unethical users.
7. If we are using AWS oe Azure, the you can store the api keys in secret manager.
