## Application Programming Interface

It is a bridge between the servers to exchange info back and forth. Security Keys or Authorisation keys are used to handle authorised requests and responses, which will prevent security breaches nd info theft.

Requests is a package in python that will help to perform API operations. If you want something from a website, then we will use GET, if you want to send something like some input, the we use POST

```python
import requests
```

## GET METHOD

```python
url='https://wttr.in/'
#In params, you are saying that I want the response in the json format. This parameter is exclusive to this API. Not part of requests.
response=requests.get(url, params={'format':'j1'})
print(f"Response: {response}")
print(f"Type of Response: {type(response)}")
print(f"Response in text Format: {response.text}")
print(f"Response in JSON Format: {response.json()}")
```

```python
response.status_code
```

## POST METHOD

```python
url="https://httpbin.org/post"
#JSON body of the API
my_data={"name":"Veena",
        "purpose":"Testing the post Method"}

response = requests.post(url,json=my_data)
print(f"response: {response}")
print(f"response in JSON: {response.json()}")
```

==================== REQUESTS – FINAL CHEAT SHEET ====================

CORE IDEA:
requests → used to send HTTP requests (GET, POST) to APIs

---------------------------------------------------------------------

GET vs POST (MAIN DIFFERENCE)

GET  → FETCH data
       → data goes in URL (params)

POST → SEND data
       → data goes in BODY (json / data)

---------------------------------------------------------------------

WHERE EACH PARAMETER IS USED

🔹 USED IN BOTH (GET & POST)

headers → auth (token, API key), content-type  
timeout → stop slow requests  
auth    → username/password login  
cookies → session handling  
verify  → SSL check (False = unsafe ⚠️)  
stream  → handle large responses  
proxies → send via proxy  

---------------------------------------------------------------------

🔹 MOSTLY USED IN GET

params → sends data in URL  
         example: ?city=London  

✔️ Can also be used in POST (less common)

---------------------------------------------------------------------

🔹 MOSTLY USED IN POST (IMPORTANT)

json → sends JSON data in BODY ✅ (MOST USED in APIs)

data → sends form data in BODY  
        (application/x-www-form-urlencoded)

files → upload files  

❌ Not used in GET (generally ignored)

---------------------------------------------------------------------

RESPONSE (WORKS SAME FOR BOTH)

status_code → 200 = success  
text        → raw string response  
json()      → converts to Python dict ✅  

---------------------------------------------------------------------

KEY DIFFERENCE SUMMARY

GET  → params → URL  
POST → json   → BODY (preferred)  

data → BODY (form, older style)

---------------------------------------------------------------------

MEMORY TRICK 🧠

GET  = "Give me data" (URL)  
POST = "Take this data" (BODY)  

---------------------------------------------------------------------

IMPORTANT NOTE

✔ Same parameters exist for both  
❗ But usage differs:

- params → mainly GET  
- json/data → mainly POST  

=====================================================================

## Handling the http error using try and except

```python
try:
    url="https://wtt908098r.in/"
    respomse = requests.get(url,params={"format":"j1"})
    print(f"respomse:: {respomse}")
except Exception as e:
    print(f"Exception occured:: {e}")

    
```
