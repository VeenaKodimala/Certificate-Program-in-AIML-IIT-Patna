```python
import pandas as pd
from matplotlib import pyplot as plt
from plotly import express as py
import seaborn as sb
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
df.head()
```

#### Task 1: Distribution Analysis Using Histograms

```python
# Histogram for total_bill
fig,ax = plt.subplots(1,2,figsize=(10,5))

ax[0].hist(df['total_bill'],bins=5)
ax[0].set_xlabel("Range of Total bill")
ax[0].set_ylabel("Total Bill")

ax[1].hist(df['total_bill'],bins=15)
ax[1].set_xlabel("Range of Total bill")
ax[1].set_ylabel("Total Bill")
```

The distribution of total_bill histogram is left-skewed.

By changing the no.of bins , by increasing them, the ranges of division will be less and the data can be plotted in a more detailed way, we understand it more clearly.
