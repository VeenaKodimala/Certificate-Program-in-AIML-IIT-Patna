```python
import pandas as pd
from matplotlib import pyplot as plt
from plotly import express as py
import seaborn as sb
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
df.head()
import plotly.express as ply
```

#### Task 1: Distribution Analysis Using Histograms

```python
# Histogram for total_bill
fig,ax = plt.subplots(1,2,figsize=(10,5))

ax[0].hist(df['total_bill'],bins=5)
ax[0].set_xlabel("Range of Total bill")
ax[0].set_ylabel("Total Bill")

ax[1].hist(df['total_bill'],bins=15)
ax[1].set_xlabel("Range of Total bill")
ax[1].set_ylabel("Total Bill")
```

The distribution of total_bill histogram is left-skewed.

By changing the no.of bins , by increasing them, the ranges of division will be less and the data can be plotted in a more detailed way, we understand it more clearly.

#### Task 2: Outlier Detection Using Box Plots

```python
sb.boxenplot(x='day',y='total_bill',data=df,palette='pastel')
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Day wise split of Total Bill")

#Outlier Detection
days=['Thur','Fri','Sat','Sun']

for i in days:
    q1 = df[df['day'] == i]['total_bill'].quantile(0.25)
    q3 = df[df['day'] == i]['total_bill'].quantile(0/75)
    print(f"q1: {q1}, q3: {q3}")
    iqr = q3-q1
    print(f"iqr: {iqr}")
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 - 1.5 * iqr
    
    outliers = df[((df['total_bill'] > upper_bound) | (df['total_bill'] < lower_bound)) & (df['day'] == i)]

    print(f"Outliers of {i} : {outliers}")



    
```
