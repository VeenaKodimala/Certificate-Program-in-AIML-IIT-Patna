## difference between pyplot and subplot()
1. Two key roles of pyplot is to create objects(axis and figure) and quick plotting. Using pyplot we will be able to create graphs easily w/o worring abt axis and all. Matplotlib will automatically creates figure and axis when we do plt.plot(). But we will not have control on it. Like if we try to plot two graphs like in cell39, both will be plotted in same axis, we will not have control on it. We will be confused which is what. That is where we can use subplots()
2. the subplots() method of pyplot creates the objects like figure and axis. Rest of the plotting is of object-oriented approach using axws(ax.plot) and not state based(plt.plot). So pyplot is just enabling you via subplots() to implement object-oriented approach. Here there will be only one 1 figure but when we plot multiplegraphs, each will have it's own axes. Syntax: subplots(rows,columns).If 1,2 then we will have one row of graphs with 2 columns, meaning 1 graph per column.

## Bar Chart

```python
### VERTICAL BARS
departments = ["CSE","ECE","ME","CE"]
students = [120,100,80,60]
plt.bar(departments, students,color='violet',edgecolor='green')
plt.xlabel("Departments")
plt.ylabel("Number of Students")
plt.title("Student Distribution Across Departments")
plt.show()
```

```python
### HORIZONTAL BARS
departments = ["CSE","ECE","ME","CE"]
students = [120,100,80,60]
plt.barh(departments, students,color='violet',edgecolor='green')
plt.xlabel("Departments")
plt.ylabel("Number of Students")
plt.title("Student Distribution Across Departments")
plt.show()
```

```python
departments = ["CSE","ECE","ME"]
male = [70,60,50]
female = [50,40,30]
x= np.arange(len(departments))
width = 0.35
plt.bar(x-width/2, male, width, label="Male")
plt.bar(x + width/2, female, width, label="Female")
plt.xticks(x, departments)
plt.legend()
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt
#This will give the frequency of each element given in the input
plt.hist
marks = [45,50,55,60,65,65,70,72,75,80,85,90]
plt.hist(marks,bins=5)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Distribution of Student Marks")
plt.show()
```

## What are bin???

5. Understanding Bins
A bin represents a range of values.
Example bins:
Mathematically, if the range of data is:
Range = max(X) - min(X)
then bin width is approximately:
Bin Width = Range/Number of Bins
In Matplotlib we control bins using:
plt.hist(x,bins=5)

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(50,10,1000)
plt.hist(data, bins=20)
plt.title("Normal Distribution Histogram")
plt.show()
```

## Pie Charts

```python
import matplotlib.pyplot as plt
companies = ["A", "B", "C", "D"]
share = [40, 25, 20, 15]
explode=[0.5,0.4,0.3,0.5]
plt.pie(share, labels=companies,explode=explode,autopct='%1.1f%%')
plt.show()
```

```python
categories=["Research","Marketing","Infrastructure", "Salaries"]
budget = [35,25,20,20]
explode=[0.1,0.1,0.1,0.1]
plt.pie(budget, labels=categories,explode=explode, autopct='%1.1f%%')
plt.title("Company Budget Allocation")
plt.show()
```
