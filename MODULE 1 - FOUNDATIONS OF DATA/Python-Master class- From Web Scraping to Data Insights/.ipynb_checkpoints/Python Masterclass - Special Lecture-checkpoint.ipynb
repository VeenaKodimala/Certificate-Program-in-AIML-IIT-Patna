```python
from bs4 import BeautifulSoup
# This is exactly how a real website looks inside
html ='''
<div class="product">
<h2 class="name">Air Jordan 1</h2>
<span class="price">$180</span>
<span class='brand'>Nike</span>
</div>
'''
soup=BeautifulSoup (html, "html.parser")
print(f"soup:: {soup}")
#Thisl will give the first occurance of h2 of class name => nmae
name=soup.find("h2", class_="name")
print(f"name:: {name}")
print(f"name.text:: {name.text}")
brand=soup 
```

```python
html ='''
<div class='product'><h2 class='name'>Nike</h2><span class='price'>Rs. 700</span></div>
<div class='product'><h2 class='name'>Adidas</h2><span class='price'>Rs. 800</span></div>
<div class='product'><h2 class='name'>Spark</h2><span class='price'>Rs. 1200</span></div>
'''

soup = BeautifulSoup(html,"html.parser")
#This provide a list of all the elements with div and classname => product
products=soup.find_all("div",class_="product")
print(products)
sneakers=[]
for product in products:
    sneker = {
        'Product Name':product.find("h2",class_="name").text,
    'price':product.find("span",class_="price").text
    }
    sneakers.append(sneker)

print(sneakers)    
```

```python
import requests
url='https://books.toscrape.com/'
response=requests.get(url)
print(response.status_code)
soup=BeautifulSoup(response.content, 'html.parser')
#Each product is in <article class='product_pod'>
products=soup.find_all('article', class_='product_pod')
#print(products)
for product in products [:3]:
    #First 3 only
    name = product.find("h3").find("a")["title"]
    price =product.find("p", class_="price_color").text
    print(name,price)
```

```python
import pandas as pd
import numpy as np
raw_data = [
{"name": "Air Jordan 1","price":"$180","rating":"4.5"},
{"name": "Nike Dunk Low","price":"$118", "rating": ""},
{"name": "Yeezy 350","price":"$220","rating":"4.8"},
{"name": "New Balance 550", "price": "SN/A", "rating": "4.2"}
]
df= pd.DataFrame (raw_data)
print('BEFORE cleaning:')
print(df)
```

```python
#df['price']=df['price'].str.replace('$','',regex=False).str.strip()
#df['price']=pd.to_numeric(df['price'],errors='coerce')
df["rating"] = df["rating"].replace("", np.nan)
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df['price']=df['price'].fillna(df['price'].median())
df['rating']=df['rating'].fillna(round(df['rating'].median(),1))
print(f"After cleaning: {df}")
```
