## Context Manager
Usually, when we open any files or any resource, we have to close it. In java this has to be provided in finally block. So irrespective of success and failed utilisation of the resource, the resource has to be closed on mandatory. In python, this will be taken care by Context Managers.

Eg: "with" keyword is a context manager.

File handling is a crucial part of any web application or data processing task. Python has several functions for creating, reading, updating, and deleting files.
The key function for working with files in Python is the open( ) function. It takes two parameters: filename and mode.
Modes
1. "r" - Read - Default value. Opens a file for reading, error if the file does not exist
2. "a" - Append - Opens a file for appending, creates the file if it does not exist. When we open a file in write mode, the old data in that file(if old file), will be appended at the end.
3. "w" - Write - Opens a file for writing, creates the file if it does not exist. When we open a file in write mode, the old data in that file(if old file), will be overridden.
4. "x" - Create - Creates the specified file, returns an error if the file exists
5. "t" - Text - Default value. Text mode
6. "b" - Binary - Binary mode (e.g. images)

```python
#f is an object of context manager "with"
with open("FileMgmt1.txt","w") as cntxtmgr:
    cntxtmgr.write("Hello, file creation via python code implementing context manager")
```

```python
#f is an object of context manager "with"
with open("FileMgmt1.txt","w") as cntxtmgr:
    cntxtmgr.write("Hello again, file opened in write mode, so overwrote the existing content present before")
```

```python
#f is an object of context manager "with"
with open("FileMgmt1.txt","a") as cntxtmgr:
    cntxtmgr.write("Hello again, file opened in append mode, adding additional data to the existing content")
```

```python
#Creating json object to fill the data in a file using json obj
# Difference between json.dump and write() is, if we use json.dump() to write a python dictionsry as a json object into file, it will 
#automatically takes care of the formatting, and will be included in file as "json" object, which will later 
#be used for json.load() to load json object from file. Where as for write(), it will include the provided dict as a string into file.
#This will lead to error in json.load() function.
import json

studentData = {"name":"Xoya Rhey","major subject":"Ccomputer Science","age":25}

with open("studentdata.json","w") as cntxtmgr:
    #dumping the data in studentData in studentdata.json as a json file
    json.dump(studentData,cntxtmgr)

with open("studentdata.json","r") as cntxtmgr:
    load_data = json.load(cntxtmgr)
    print(load_data)
    
    
```

Easy
1. Write a Python program to create a file named notes.txt and write "Learning Python is fun!" into it.
2. Read the first 5 characters of notes.txt and print them.


```python
with open("notes.txt","w") as cntxtmgr:
    cntxtmgr.write("Learning Python is fun!")

with open ("notes.txt","r") as cntxtmgr:
    data = cntxtmgr.read(5)
    print(data)    
```

Medium
3. Write a function count_words (filename) that returns the total number of words in a given text file.
4. Create a program that reads a file and writes its content into a new file named backup.txt, but in all uppercase letters.

```python
def count_words (filename):
    with open(filename,"r") as cntxtmgr:
        data =  cntxtmgr.read()
        #count no.of words
        print(f"total no.of words: {len(data.split())}")
        #count total no.of chars
        print(f"total no.of chars: {len(data)}")

count_words("notes.txt")   
```

```python
def transferContent (srcfile,destfile):
    try:
        data = ""
        with open(srcfile,"r") as cntxtmgr:
            data =  cntxtmgr.read()

        with open(destfile,"w") as cntxtmgr:
            cntxtmgr.write(data.upper())

        return "Successfully transferred the content."

    except FileNotFoundError:
        return "File Not Found"


rtnMsg = transferContent("notes.txt","notesCopy.txt")   
print(rtnMsg)
```

Hard
5. Given a CSV-like text file scores.txt with content:
Alice, 85
Bob, 92
Charlie, 78
write a script to calulate the average score.
6. Write a robust code to open a file provided bu user input. If the file is not found, print a user-friendly message.

```python
with open("score.txt","w") as f:
    f.write("Alice, 85\nBob, 92\nCharlie, 78")

scoresum = 0
count = 0
with open("score.txt","r") as f:     
    for line in f:
        
        count = count + 1
        scoresum = scoresum + int(line.split()[1])
        
print(scoresum) 
print(count)
print(scoresum/count)             
       
```
