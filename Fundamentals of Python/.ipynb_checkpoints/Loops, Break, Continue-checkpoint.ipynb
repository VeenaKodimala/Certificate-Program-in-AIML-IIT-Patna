## LOOPS(IF, WHILE), BREAK, CONTINUE


```python
"""
Syntax of for loop:

for i in range(start,stop,increment):
by defaulyt the increment=1, start=0, but the stop is a mandatory param to provide.
If only one value is provided in range then it is considered for stop, if 2 then start and stop. If 3 then only it will consider for
start, stop and increment.
"""

#Example
for i in range(5):
    print(f"i::{i}")

for i in range(5,31,5):
    print(f"i::{i}")
```

### EDGE CASES IN FOR LOOP

```python
for i in range(5,5):
    print("Never printing line")
print("Only this line will be printed")    

for i in range(10,0,-2):
    print(f"i::{i}")

for i in range(10,0,2):
    print(f"i::{i}")    
```

```python
forList = [34,67,98,242,7565,7565,98,98]

for score in forList:
    print(f"score:: {score}")

scentence="My name is Zoya Rheya"
for letter in scentence:
    print(f"letter :: {letter}")
```

We can use for loop when we know specific range, but if we do not have a apecific range but condition when to stop, we 
can use while loop.

```python
i=0
while i<=5 :
    print(i)
    i = i + 1
```

### Break Statement - It is basically stop the loop and come out of it completely

```python
i=0
books=["b1","b2","b3","b4","b4","b5","b6","b7","b8","b9","b10"]
while i < len(books):
    print("current book:: ", books[i])
    if(books[i] == 'b5'):
        break
    i = i+1    
print("for loop.....")
for book in books:
    print("current book:: ", book)
    if(book == 'b5'):
        break
    
```

### Continue Statement - It is basically stop the current loop and continue with the next iteration of the loop.

```python
books=["b1","b2","b3","b4","b5","b6","b7","b8","b9","b10"]
for book in books:    
    if(book == 'b5'):
        continue
    print("current book:: ", book)    
```

```python
count=9
while count < 15:
    print("count b4 increment::",count)
    count = count + 1
    if(count == 10):
        continue
    print("count::",count)
    
```
