### Functions syntax

```python
def func_name(args1,args2):
    print(f"args1:: {args1} args2 :: {args2}")
```

func_name(2,3)

### Positional Parameters
1. Here the function call and func def will have same order of args and params.

```python
def sampleFunc(args1,args2):
    print(f"args1:: {args1} args2 :: {args2}")

sampleFunc(10,23)    
```

### Keyword Arguments
1. Here the function call will assign the params using args names

```python
def sampleFunc(args1,args2):
    print(f"args1:: {args1} args2 :: {args2}")

sampleFunc(args2=10,args1=23)  
```

### Combo of positional and keyword arguments

```python
def sampleFunc(args1,args2,args3=890):
    print(f"args1:: {args1} args2 :: {args2} args3 :: {args3}")

sampleFunc(args2=10,args1=23)    
```

### Ordering of positional and keyword argumets
1. Always we need to order the positional first and then order keyword next. So if you are deciding to use both combo of keyword and positional from func call, then the first parameter can never be "Key word". If the first para is key word, thwn the rest should also be keyword, else it will throw error. So the order is important, 1st - positional, then rest. And you cannot do like this - positional,keuword,positional ...

```python
def sampleFunc(args1,args2,args3):
    print(f"args1:: {args1} args2 :: {args2} args3 :: {args3}")

sampleFunc(args2=10,args1=23,890) 
```

```python
def sampleFunc(args1,args2,args3):
    print(f"args1:: {args1} args2 :: {args2} args3 :: {args3}")

sampleFunc(10,args2=23,args3=890)
```

### Variable no.of args
To be used when you are not sure how many args you will pass, and dont know the definite count of the params.

```python
def calc(*args):
    print(f"sum of a and b:: {args[0]+args[1]} difference of b and c :: {args[1]-args[2]} product of a and c :: {args[0]*args[2]}")

calc(10,23,890) 
```

### Variable keyword params

```python
def kw_args(**args):
    for key,val in args.items():
        print(f"key:: {key} value :: {val}")


kw_args(apple=100,guava=23,hanana=45)
```

### Lambda Functions

```python
evan_lamba = lambda a: a%2==0
lst = [5,4,7,9,8,5,4,4,6,7,8]
for val in lst:
    print(evan_lamba(val))
```

### Recurssion

```python
def fact(n):
    if(n == 1):
        return n
    else:
        return n*fact(n-1)


print(fact(5))
```

1. odd or even

```python
def oddoreven(n):
    if(n%2==0):
        print("even")
    else:
        print("odd")

oddoreven(9)
```

2. Calculate the area of circle/rectangle/square - are(shape,**dimentions)

```python
import math

def areaShape(shape,**dimensions):
    if(shape == 'circle'):
        return (math.pi * (dimensions['radius']**2))
    elif(shape == 'rectangle'):
        return dimensions['length'] * dimensions['width']    
    else:
        print("unknown shape")

print(areaShape('circle',radius=2))
print(areaShape('rectangle',length=2,width=4))
```
