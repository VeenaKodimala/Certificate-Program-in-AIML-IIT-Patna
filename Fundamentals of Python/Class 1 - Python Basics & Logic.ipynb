## Logical Operators

```python
has_credits = False
has_api_key = True

can_run_api = has_credits and has_api_key 
have_partial_Api_details = has_credits or has_api_key
need_topup = not has_credits

print("Can we run the api:", can_run_api)
print("Do we any details of the api:",have_partial_Api_details)
print("Do we need credits top up:",need_topup)
```

## AI Response Filter Task

```python
try:
    confidence_Score=int(input("Enter your confidence score:"))
    if(confidence_Score >= 80):
        print("High confidence - Accept")
    elif(confidence_Score >= 50):
        print("Medium - Review")
    else:
        print("Low - Reject")

except ValueError:
    print("Please enter valid number")

finally:
    print("End of the task")
```

## Scenario: You need a standard way to log Al interactions in your application.
Requirements:
1. Write a function called format_api_response that takes three parameters: model, prompt, and tokens
2. Inside the function, create a dictionary with the following keys:
"model" the model name.
"input": the prompt text.
"token count" the number of tokens.
3. The function should return this dictionary.
4. Test your function by calling it with sample data (e.g., "pt-4", "What is Python?", 150).
5. Print the resulting dictionary to the console.

```python
def logAI(model,input,tokens):
    log = {"model":model,"input":input,"token count":tokens}
    return log

logs = logAI("pt-4","What is Python",150)    
print("AI Logs:: ",logs)
```
