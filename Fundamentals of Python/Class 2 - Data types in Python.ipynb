# DATA TYPES IN PYTHON
1. Integer - all the +ve and -ve numbers. Eg: 6,-9
2. Float - numbers with decimal points. Eg: 8.9
3. string - Eg: "apple", "1","-56"
4. Complex - quadratic equations(x+yj). Eg: 2+6j
5. boolean - TRUE/FALSE.
6. None



We use single, double, triple quotes to store a string.
When are any single quotes in your sentence, the we will use double quotes to enclose the string. If
your scentence has both double and triple quotes, then we use triple quotes to enclose the string.

Eg: 1. name='Veena'
2. Veena_frns = "Veena's Friend"
3. VeenaJob = '''Veena's Friend"s job'''


```python
# Integer
x = 10
y = -78
print("type(x):: ", type(x)," type(y):: ",type(y))

#float
a = 8.90
print("type(a):: ", type(a))

#string
b="word"
c="67"
d="-67"
name='Veena'
Veena_frns = "Veena's Friend"
VeenaJob = '''Veena's Friend"s job'''
print("type(b):: ", type(b)," type(c):: ",type(c)," type(d):: ",type(d))
print("name:: ", name," Veena_frns:: ",Veena_frns," VeenaJob:: ",VeenaJob)

#Complex
e=complex(4,5)
f=complex(6)
g=complex()

print("e::",e)
print("f::",f)
print("g::",g)

print("e's real part::",e.real)
print("e's img part::",e.real)
print("f's real part::",f.real)
print("f's real part::",f.real)
print("g's real part::",g.real)
print("g's real part::",g.real)
print("type(e)::",type(e))
print("type(g)::",type(g))

#boolean
h=10
i="10"
j=90
k=90
print("h==i::", h==i)
print("j==k::", j==k)

#None

l=None
print("type(l):::",type(l))
```
