3. Given two lists keys = ['a', 'b', 'c'] and values = [1, 2, 31, create a dictionary using a loop or comprehension.
4. Write a function that takes a dictionary and returns a new dictionary with keys and values swapped.


```python
def swapDict(keylst,vallist):
    swappedDict = {v:k for k,v in zip(keylst,vallist)}
    

    print(swappedDict)
    return swappedDict
    
keys=['a','b','c']
values=[1,2,31]

dictionaryApproach1 = {}
dictionaryApproach2 = {}

#Approach - 1 --> Old way
for i in range(0, len(keys)):
    dictionaryApproach1[keys[i]]=values[i]

#Approach - 2
for key,value in zip(keys,values):
    dictionaryApproach2[key]=value

#Approach - 3 --> Dictionary comprehension
# Zip function basically creates multiple itatables form different entities.
# the no.of itratables created is equal to the no.of entities zipped.
# the entities inside the zip can be duplicated. eg:  zip(keys,keys)
# zip can be applied for different length entities, but the itratables stop at the shortest length.d 
dictionaryApproach3 = {k:v for k,v in zip(keys,values)}

print("dictionaryApproach1::",dictionaryApproach1)  
#the below printing approach used f strings.
print(f"dictionaryApproach2::{dictionaryApproach2}")  
print("dictionaryApproach3::",dictionaryApproach3)  

swappedDict = swapDict(keys,values)
print("swappedDict::",swappedDict)  

```

5. Given a string text = "apple banana apple cherry banana apple", create a dictionary that counts the frequency of each word.
6. Merge two dictionaries d1 = {'a': 10, 'b': 20 and d2 {'b': 30, 'c': 40} such that if a key exists in both, their values are added together.

```python
def mergeDict(d1,d2):
    mergedDictn = d1.copy()
    print(mergedDictn) 

    for key,value in d2.items():
        mergedDictn[key] = mergedDictn.get(key,0)+value 

    print("mergedDictnmergedDictn:: ",mergedDictn) 
    return mergedDictn


text = "apple banana apple cherry banana apple"
print(text)
text.split()
fruits={}

for fruit in text.split():
    fruits[fruit] = fruits.get(fruit,0)+1

print(fruits)
d1 = {'a': 10, 'b': 20}
d2 = {'b': 30, 'c': 40}
print("Merged dict form mergeDict:: ",mergeDict(d1,d2))
```
