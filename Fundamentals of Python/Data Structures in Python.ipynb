#### Basic Operations of List
1. append - add an element to the list, it will accept duplicate as weel, coz, lists allow duplication.
2. insert - If you want to insert the element in a desired position.
3. extend - To join multiple lists. This will allow joining same array multiple times.
4. remove - To remove an element from list, basically the first occurance. It will throw a value error if the specified element is not present in the list.
5. pop - To pop the last index of the list
6. del - delete a desired element from list
7. clear - Clearing the entire list
8. index() - to get the index of an element
9. count - To get the no.of occurences of an element.
10. in and not in  - To check if an element is present in a List
11. sort - To sort a lsit in ascending order
12. sort(reverse=True) - To sort a lsit descending order
13. sorted() - If you want t o sort a list but do not touch the existing list.
14. len() - to know the lwngth of list.

```python
fruits = ['apple','orange','banana','grapes']
veggies = ['Beetroot','Carrot','Beans']
#Add an element at last
fruits.append('Guava')
print(fruits)

#Insert the element at a desired position - in insert the frst index from last will 
fruits.insert(3,'Parsimon')
fruits.insert(-1,'Peach')
print(fruits)

#Join two lists - extend
#This extends will work in joining a list and tuple as well
fruits.extend(veggies)
print(fruits)
veggies.extend(veggies)
print(veggies)

#remove an element
veggies.remove('Beans')
print(veggies)

#pop - to remove the last element of the lsit(if no index is specified), otherwise the elemeent in the index specified.
#iN POP, if we mention -1 to remove, it will remove the last elemnt from last.
print(f"removing {fruits.pop()} from fruits -- {fruits}")
print(f"removing {fruits.pop(-1)} from fruits -- {fruits}") 

#del - delete an element from list
del veggies[2]
print(veggies)

#clear - clearing the entire list
veggies.clear()
print(f"Cleared the entire veggies list: {veggies}")

#index - to find teh index of the element
print(f"Index of Banana : {fruits.index('banana')}")

fruits.append('Beetroot')
print(f"No.of occurenaces of Beetroot: {fruits.count('Beetroot')}")

print(f"Is Okra present in list: {'Okra' in fruits}")
print(f"Okra present in list: {'Okra' not in fruits}")

#Assigning a value to an index
fruits[3] = 'Veena'
fruits[-3] = 'Zoya'
print(fruits)
fruits.sort()
print("ascending order for str list:",fruits)

numbers = [34,78,34,876,123,678,3457,65]
numbers.sort()
print(f"Numbers in ascending order: {numbers}")
numbers.sort(reverse=True)
print(f"numbers in descending order: {numbers}")

numCopy = sorted(numbers)
print(f"numCopy:: {numCopy}")
print(f"length of numbers: {len(numCopy)}")

# Enumerate to get both tne index and value at the same time.
for i in enumerate(numbers):
    print(f"i:: {i}")

for index,value in enumerate(numbers):
    print(f"index: {index}, value: {value}")


cordinates=[23,45,67,(56,67,66),98]
print(cordinates)
cordinates[3][1] = 'Veena'
print(cordinates)
```

#### List Slicing
syntax: listname[start : end : step] - end index is excluded

```python
objs = ['apple', 'orange', 'banana', 'Parsimon', 'grapes', 'Peach', 'Guava', 'Beetroot', 'Carrot']
obj1 = objs[2:5]
print(obj1)
obj2 = objs[4:]
print(obj2)
obj3 = objs[:5]
print(obj3)
obj4 = objs[:]
print(obj4)
obj5=objs[-5:-1]
print(obj5)
obj6=objs[:-5]
print(obj6)
```

```python
#Emoty set
a = set()
print(a)
lstnNum = [1,2,3,3,4,4,5,5,6]
setNum = set(lstnNum)
setNum.add(0)
print(setNum)

greet = "Hello"
print(set(greet))
```

```python
a = {1,2,3,4,5,6}
b = {4,5,6,7,8,9}

# Union of sets
print(a | b)
print(a.union(b))

# Intersection of sets
print(a & b)
print(a.intersection(b))

# Sutraction of sets
print(a-b)
print(a.difference(b))
print(b-a)
print(b.difference(a))

#Subset - To check if the elements of first is present in second
#Superset - To check if the elements of second is present in first
c={1,2,3}
d={1,2,3,4,5}    
print(c.issubset(d))
print(c.issuperset(d))
c={1,2,3}
d={1,2,3,4,5}    
e={8,9,0}
print(d.issubset(c))
print(c.issuperset(d))
print(c.isdisjoint(e))
print(c.isdisjoint(d))
```
