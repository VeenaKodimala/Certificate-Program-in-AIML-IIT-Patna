## Context Manager
Usually, when we open any files or any resource, we have to close it. In java this has to be provided in finally block. So irrespective of success and failed utilisation of the resource, the resource has to be closed on mandatory. In python, this will be taken care by Context Managers.

Eg: "with" keyword is a context manager.

File handling is a crucial part of any web application or data processing task. Python has several functions for creating, reading, updating, and deleting files.
The key function for working with files in Python is the open( ) function. It takes two parameters: filename and mode.
Modes
1. "r" - Read - Default value. Opens a file for reading, error if the file does not exist
2. "a" - Append - Opens a file for appending, creates the file if it does not exist. When we open a file in write mode, the old data in that file(if old file), will be appended at the end.
3. "w" - Write - Opens a file for writing, creates the file if it does not exist. When we open a file in write mode, the old data in that file(if old file), will be overridden.
4. "x" - Create - Creates the specified file, returns an error if the file exists
5. "t" - Text - Default value. Text mode
6. "b" - Binary - Binary mode (e.g. images)

```python
#f is an object of context manager "with"
with open("FileMgmt1.txt","w") as cntxtmgr:
    cntxtmgr.write("Hello, file creation via python code implementing context manager")
```

```python
#f is an object of context manager "with"
with open("FileMgmt1.txt","w") as cntxtmgr:
    cntxtmgr.write("Hello again, file opened in write mode, so overwrote the existing content present before")
```

```python
#f is an object of context manager "with"
with open("FileMgmt1.txt","a") as cntxtmgr:
    cntxtmgr.write("Hello again, file opened in append mode, adding additional data to the existing content")
```

```python
#Creating json object to fill the data in a file using json obj
import json

studentData = {"name":"Xoya Rhey","major subject":"Ccomputer Science","age":25}

with open("studentdata.json","w") as cntxtmgr:
    #dumping the data in studentData in studentdata.json as a json file
    json.dump(studentData,cntxtmgr)

with open("studentdata.json","r") as cntxtmgr:
    load_data = json.load(cntxtmgr)
    print(load_data)
    
    
```
