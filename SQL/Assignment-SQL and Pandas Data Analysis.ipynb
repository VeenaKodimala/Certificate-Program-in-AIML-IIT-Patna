## Task 1: Create Database

```python
import pandas as pd
import sqlite3
import random
```

```python
conn = sqlite3.connect('employeeDB')
cursor=conn.cursor()
conn.commit
```

```python
cursor.close()
conn.close()
```

```python
query="CREATE TABLE IF NOT EXISTS EMPLOYEES( \
emp_id INTEGER, \
name TEXT, \
department TEXT, \
salary REAL, \
year_exp INTEGER, \
performance_score INTEGER \
)"
departments = ["Engineering", "Sales", "Marketing", "HR", "Finance"]
empData=[]
cursor.execute(query)

for i in range (1,41):
    name=f"Employee_{i}"
    department = random.choice(departments)
    salary = round(random.uniform(50000,150000),2)
    year_exp=random.randint(0,15)
    score=round(random.uniform(1,5),1)

    empData.append((i,name,department,salary,year_exp,score))

#print(f"empData:: {empData}")

cursor.execute("DELETE FROM EMPLOYEES")
query="INSERT INTO EMPLOYEES (emp_id,name,department,salary,year_exp,performance_score) \
values(?,?,?,?,?,?)"

cursor.executemany(query,empData)


cursor.execute("SELECT * FROM EMPLOYEES")     
cursor.fetchall()
    
```

## Task 2: SQL Queries

```python
query1 = "select name, department, salary, performance_score from EMPLOYEES WHERE performance_score >= 4.0 AND year_exp >= 3 \
ORDER BY performance_score LIMIT 15"

cursor.execute(query1)
print("--------------------------QUERY-1----------------------")
print(cursor.fetchall())

query2 = "select * from EMPLOYEES where salary BETWEEN 70000 AND 110000 AND department in ('Engineering','Sales') order by salary desc"
cursor.execute(query2)
print("--------------------------QUERY-2-----------------------")
print(cursor.fetchall())

query3 = "select count(*),round(avg(salary),2) as avg_salary from EMPLOYEES GROUP BY department order by avg_salary desc"
cursor.execute(query3)
print("--------------------------QUERY-3-----------------------")
print(cursor.fetchall())
```

## Which is better for datasets too large for memory? Why?

SQL and Pandas are both powerful for data analysis, but their syntax and use cases differ significantly. In SQL, conditions use = for equality and AND for combining filters, whereas Pandas uses == and & inside .loc[]. Sorting in SQL is done with ORDER BY, while in Pandas it is handled using .sort_values(). SQL queries are declarative—you describe what you want—while Pandas is more procedural, showing how to do it step by step.

SQL is best suited for working directly with databases, especially when handling structured data stored in relational systems. It excels at joins, aggregations, and filtering large datasets efficiently on the server side. Pandas, on the other hand, is ideal for in-memory data analysis, quick transformations, and exploratory work in Python, offering more flexibility for custom logic and integration with libraries like NumPy and Matplotlib.

For datasets too large to fit into memory, SQL is generally better. This is because databases process data on disk and use optimized query engines, indexing, and execution plans, whereas Pandas loads data into RAM, which can lead to memory limitations and performance issues.
