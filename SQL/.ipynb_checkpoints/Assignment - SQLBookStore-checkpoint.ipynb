# 📚 Task 1: Database Design and Table Creation

```python
import sqlite3
import pandas as pd
```

```python
bookstoreDB = sqlite3.connect('bookstore.db')
cursor = bookstoreDB.cursor()
bookstoreDB.commit()
```

```python
cursor.close()
bookstoreDB.close()
```

```python
cursor.execute("CREATE TABLE IF NOT EXISTS BOOKSTBL(book_id INTEGER PRIMARY KEY AUTOINCREMENT, \
title TEXT Not Null,author TEXT Not Null, \
price REAL Not Null, \
stock_quantity INTEGER Default 0)")

cursor.execute("CREATE TABLE IF NOT EXISTS CustomerInfo(customer_id INTEGER Primary Key Autoincrement, \
name TEXT Not Null, \
email TEXT Unique Not Null, \
city TEXT, \
join_date TEXT)")


cursor.execute("CREATE TABLE IF NOT EXISTS Orders(order_id INTEGER Primary Key Autoincrement, \
customer_id INTEGER , \
book_id INTEGER, \
quantity INTEGER Not Null, \
order_date TEXT Not Null, \
total_amount REAL, FOREIGN KEY(customer_id) references CustomerInfo(customer_id),FOREIGN KEY(book_id) references BOOKSTBL(book_id))")


query="SELECT name FROM sqlite_master where type='table'"
#The cursor will now hold the result of the query, whcih can be fetched using fetch methods.
tables=cursor.execute(query)
tables=cursor.fetchall()

for table in tables:
    tablename=table[0]
    query=f"PRAGMA table_info('{tablename}')"
    cursor.execute(query)
    colnames=[col[0] for col in cursor.description]

    print(f"Schema of table: {tablename}")
    print(colnames)
    for row in cursor.fetchall():
        print(row)
```

## PART - B : COMPLEX QUERIES

```python
query="SELECT * FROM CUSTOMERINFO WHERE CITY='Mumbai'"
cursor.execute(query)
colNames = [col[0] for col in cursor.description]
print(colNames)
print(cursor.fetchall())

query="SELECT * FROM BOOKSTBL WHERE price > 800 AND stock_quantity > 10"
cursor.execute(query)
colNames = [col[0] for col in cursor.description]
print(colNames)
print(cursor.fetchall())

query="SELECT count(*) FROM orders"
cursor.execute(query)
colNames = [col[0] for col in cursor.description]
print(colNames)
print(cursor.fetchall())

query="SELECT count(*) as orderCount,a.customer_id,b.name from orders a inner join customerinfo b \
on a.customer_id=b.customer_id group by a.customer_id order by orderCount desc"
cursor.execute(query)
colNames = [col[0] for col in cursor.description]
print(colNames)
print(cursor.fetchall())
```

## TASk 3 : PANDAS INTEGRATION AND ANALYSUS

### PART - A

```python
customerInfoDF = pd.read_sql("SELECT * FROM customerinfo" ,bookstoreDB)
print(customerInfoDF.head(3))

booksDF = pd.read_sql("SELECT * FROM bookstbl" ,bookstoreDB)
print(booksDF.head(3))

ordersDF = pd.read_sql("SELECT * FROM orders" ,bookstoreDB)
print(ordersDF.head(3))
```

```python
result1 = pd.merge(customerInfoDF[['customer_id','name','city']],
                   ordersDF[['customer_id','book_id','quantity','total_amount']],
                   right_on='customer_id',
                   left_on='customer_id',
                   how='inner')
result2 =  pd.merge(result1[['book_id','name','city','quantity','total_amount']],
                   booksDF[['book_id','title']],
                   right_on='book_id',
                   left_on='book_id',
                   how='inner')
result2[['name','city','quantity','total_amount','title']]
```

```python
print(ordersDF['total_amount'].mean())

result = pd.merge(customerInfoDF[['customer_id','city']],
                   ordersDF,
                   right_on='customer_id',
                   left_on='customer_id',
                   how='inner')
print(result.groupby('city')['order_id'].count().sort_values(ascending=False).reset_index(name='Total Orders'))

result = pd.merge(booksDF[['book_id','title']],
                   ordersDF,
                   right_on='book_id',
                   left_on='book_id',
                   how='inner')
result2 = result.groupby('title')['quantity'].sum().sort_values(ascending=False).reset_index(name='Total order quantity').head(1)
print(f"Most ordered book : {result2.loc[0]['title']}")
```

### PART - B

```python
discounts_data = {
    'book_id': [1, 2, 3, 4, 5],
    'discount_percent': [10, 15, 5, 20, 12]
}

discountsDF = pd.DataFrame(discounts_data)

discountsDF.to_sql('discounts',bookstoreDB,if_exists='replace',index=False)

query="select * from discounts"
cursor.execute(query)
colNames = [col[0] for col in cursor.description]
print(colNames)
print(cursor.fetchall())

query="select a.title,b.discount_percent from bookstbl a inner join discounts b on a.book_id=b.book_id"
cursor.execute(query)
colNames = [col[0] for col in cursor.description]
print(colNames)
print(cursor.fetchall())
```
