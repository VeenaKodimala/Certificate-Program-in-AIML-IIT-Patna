## Built In functions of NumPy
Arrays are homogeneous - all elements will be of same data type, but python lists are hetrogeneous
1. Creation of array with zeros - np.zeros(rows,column) - this will create an array of 0 with specified rows and columns
2. Creation of array with same values - np.full((array shape),fill_value)

```python
#2D array
np2DArray = np.zeros((3,5))
# print(f"np2DArray:: {np2DArray}")
# print(f"np2DArray shape:: {np2DArray.shape}")

#3D array
np3DArray = np.zeros((3,5,4))
# print(f"npArray:: {np3DArray}")
# print(f"np3DArray shape:: {np3DArray.shape}")

#4D array
np4DArray = np.zeros((3,5,4,3))
#print(f"np4DArray:: {np4DArray}")
print(f"np4DArray shape:: {np4DArray.shape}")

npdynamicArray = np.full((2,3,5),5)
print(f"npdynamicArray:: {npdynamicArray}")
print(f"npdynamicArray datatype:: {npdynamicArray.dtype}")
print(f"reshaped array:: {npdynamicArray.reshape(6,5)}")


npdtypeArray = np.full((2,3,5),5,dtype='<U6')
print(f"npdtypeArray:: {npdtypeArray}")
print(f"npdtypeArray datatype:: {npdtypeArray.dtype}")

#Arrange in numpy will create an array with certain conmditions
#Syntax: np.arrange(strart,stop,step)

arrangeArr = np.arange(5,20,5)
print(f"arrangeArr:: {arrangeArr}")
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(np.sum(arr))
#15
print(arr.sum())
#15 (method syntax)
print(np.mean(arr))
#3.0
print(np.std(arr))
#1.4142135623730951 (standard deviation)
print(np.var(arr))
#2.0 (variance)
print(np.min(arr))
#1
print(np.max(arr))
#5
print(np.argmin(arr))
#0 (index of minimum)
print(np.argmax(arr))
#4 (index of maximum)
```

```python
#axis=0 --> Operation along the columns
#axis=1 --> operation along the row.

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(arr_2d.sum())
#21 (sum of all elements)
print(arr_2d.sum(axis=0))
#[579] (sum along column sums)
print(arr_2d.sum(axis=1))
#[615] (sum along row sums)
print(arr_2d.mean(axis=0))
# [2.5 3.5 4.5]
print(arr_2d.max(axis=1))
# [36]
```

## reshaping
1. Reshaping is changing the dimentions of my array, into something else, so that it fits my input. We can chnage the dimension such that the product of of dimensions given is equal to the total elements in my existing array.
2. know my one dimension is 4, but i dont wantt o calculate the second dimension, in such situations i can use -1. This -1 can be specisied for only one unkown dimension.

```python
simpleArr = np.arange(50)
print(f"simpleArr:: {simpleArr}")
reshapeArr = simpleArr.reshape(5,-1)
print(f"reshaped Array:: {reshapeArr}")
#flatten - convert any dimension array to 1-D array
print(f"flattened Array:: {reshapeArr.flatten()}")
#ravel - convert any dimension array to 1-D array. Same as flatten but faster.
print(f"Raveled Array:: {reshapeArr.ravel()}")
#convert any dimension array to 1-D array w/o using ravel and flatten
print(f"reshapind 2D to 1D Array:: {reshapeArr.reshape(-1
                                             )}")
```

```python
simpleArray = np.array([[23,24,6,7],[56,77,987,867]])
print(f"simpleArray:: {simpleArray}")
tranposedArr = simpleArray.T
print(f"tranposedArr:: {tranposedArr}")
```

```python
#Comparing Python lists and numpy arrays in terms of time.
# Create large data
import time 

size =  1000000
python_list = list(range(size))
numpy_array = np.arange(size)
# Test 1: Summing elements
# Python list
start = time.time()
total = sum(python_list)
python_time = time.time() - start
#NumPy array
start = time.time()
total = np.sum(numpy_array)
numpy_time = time.time()- start
print(f"Python list sum: {python_time:.5f} seconds")
print(f"NumPy array sum: {numpy_time:.5f} seconds")
print(f"NumPy is {python_time/numpy_time:.1f} x faster")
```

## hstack and vstack
1. Stack is known as joining arrays together
2. hstack - joining arrays horizontally. The no.of rws will be same
3. vstack - joining arrays vertically. No.of columns will bne same

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,0])
print(f"hstack::: {np.hstack((a,b))}")
print(f"vstack::: {np.vstack((a,b))}")

c = np.array([[1,2],[3,4]])
d = np.array([[6,7],[8,9]])
print(f"hstack::: {np.hstack((c,d))}")
print(f"vstack::: {np.vstack((c,d))}")
```
