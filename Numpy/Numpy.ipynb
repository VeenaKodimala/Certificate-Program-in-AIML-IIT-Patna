## NUMPY BROADCASTING

Broadcasting allows NumPy to perform element-wise operations on arrays with different shapes without copying data.

Rules:
1. If arrays have different number of dimensions, NumPy adds 1s to the LEFT of the smaller shape.
2. Shapes are compared from RIGHT → LEFT.
3. Two dimensions are compatible if:
      • they are equal
      • one of them is 1
4. If a dimension is 1, NumPy stretches (broadcasts) it to match the other dimension.
5. If neither condition is satisfied → broadcasting error.

Example:
A shape → (2,1,2)
B shape → (2,1)

Step 1: Make dimensions equal
A → (2,1,2)
B → (1,2,1)

Step 2: Compare RIGHT → LEFT
(2,1,2)
(1,2,1)

Dim3: 2 vs 1 → expand 1 → 2
Dim2: 1 vs 2 → expand 1 → 2
Dim1: 2 vs 1 → expand 1 → 2

Final broadcast shape → (2,2,2)

Memory trick:
Broadcasting works if dimensions are EQUAL or ONE of them is 1 (checked RIGHT → LEFT).

```python
################################# BROADCAST NEGATIVE EXAMPLE ###########################################
bcArray1 = np.full((2,3,5),3)
bcArray2 = np.full((3),5)
print(f"shape of bcArray1:: {np.shape(bcArray1)}, shape of bcArray2:: {np.shape(bcArray2)}")
#STEP 1 - Adding 1's to the left of the smaller shaped array.
#After changes shape:::   bcArray1 - (2,3,5), bcArray2 - (1,1,3)

#STEP - 2 - Check if the dimensions are equal or atleat one of them is 1 from right.
#Here the 1st dimensions from right are 5 and 3. None of the rule matches so this broadcast will not happen

print(f"bcArray1 + bcArray2::: {bcArray1 + bcArray2}")
                                                                                



                                                                                
                                                                                
                                                                                
```

```python
################################# BROADCAST POSITIVE EXAMPLE ###########################################

#In belo, 2=start, 20=end
PbcArray1 = np.random.randint(2,20,size=(2,3,5))
print(f"PbcArray1:: {PbcArray1}")
PbcArray2 = np.random.randint(5,50,(5))
print(f"PbcArray2:: {PbcArray2}")
print(f"shape of bcArray1:: {np.shape(PbcArray1)}, shape of bcArray2:: {np.shape(PbcArray2)}")
#STEP 1 - Adding 1's to the left of the smaller shaped array.
#After changes shape:::   bcArray1 - (2,3,5), bcArray2 - (1,1,5)

#STEP - 2 - Check if the dimensions are equal or atleat one of them is 1 from right.
#5 vs 5 -> equal - No expanding || 3 vs 1 - Expand 2nd array by 2, so 3 || 2 vs 1 - Expand 2nd array by 1, so 2
#Final broad cast size == (2,3,5)

#This broad cast will work since the shapes are brought done one final shape


print(f"bcArray1 + bcArray2::: {PbcArray1 + PbcArray2}")
                                                                                                                                                     
                                                                                
                                                                                
```

## STANDARDIZATION

Standardization is a way of changing numbers in a dataset so that the data is centered around 0 and has a similar scale. It os like shring the values range but keeping the pattern between the numbers the same.

After standardization:
• The average (mean) of the data becomes 0
• The spread of the data (standard deviation) becomes 1

Formula:
z = (x − mean) / standard_deviation

Where:
x = original value
mean = average of the data
standard_deviation = how spread out the data is
z = new standardized value

#### What this formula does:
1. (x − mean) moves the data so that the center becomes 0.
2. Dividing by standard deviation scales the data so values are not too large or too small.

#### Why we need standardization:
Sometimes different columns in data have very different ranges.

Example:
Height → 150 to 180  
Salary → 20000 to 200000

Salary numbers are much bigger than height numbers.  
Some machine learning algorithms may think salary is more important just because the numbers are larger.

Standardization fixes this by putting all columns on a similar scale.

Example:
Data = [10, 20, 30]

Mean = 20  
Standard deviation ≈ 8.16

Standardized values:
(10 − 20)/8.16 ≈ -1.22  
(20 − 20)/8.16 = 0  
(30 − 20)/8.16 ≈ 1.22

So the new values become:
[-1.22, 0, 1.22]

Key idea:
Standardization changes the scale of numbers but keeps the relative pattern of the data the same.

```python
#Problem : Standardise an array created by numpy

stddisedArr=[]
rawArr = np.array([10,20,30])
print(f"rawArr:: {rawArr}")

meanVal = np.mean(rawArr)
sdVal = round(np.std(rawArr),2);
print(f"Mean:: {meanVal}, SD:: {sdVal}")

stddisedArr = (rawArr - meanVal)/sdVal

print(stddisedArr)
    
```

## Min-Max Normalisation

Standardization doesn't have a fixed range of the standardised values. It usually makes the values with 0 as centre. But MinMax normalisation has a fixed range between 0 to 1.

Formula = (Arr - min(Arr) / ((max(arr) - min(Arr))

```python
minMaxIpArr = np.random.randint(5,20,size = (5))
print(f"minMaxIpArr:: {minMaxIpArr}")

meanVal = np.mean(minMaxIpArr)
print(f"meanVal:: {meanVal}")

minMaxArr = (minMaxIpArr - np.min(minMaxIpArr)) / (np.max(minMaxIpArr) - np.min(minMaxIpArr))
print(f"minMaxArr:: {minMaxArr}")
```
