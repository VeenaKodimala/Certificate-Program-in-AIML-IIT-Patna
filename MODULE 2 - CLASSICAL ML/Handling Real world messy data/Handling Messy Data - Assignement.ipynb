```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
df
```

```python
medianAge = df['age'].median()
print(f"Medisan of age column: {medianAge}")
df['age'] = df['age'].fillna(medianAge)
print(f"df after filling empty age cells age cells with Median: \n{df}")
```

```python
print(f"Total no.of duplicates present in dataset: {df.duplicated().sum()}")
df = df.drop_duplicates()
```

```python
print(f"df after dropping duplicates: \n{df}")
```

```python
#one-hot encoding to city
df = pd.get_dummies(df,columns=['city'],drop_first=False)
print(f"df after applying one-hot encoding to city: \n{df}")
print(f"shape of df after OHE: {df.shape}")
```

```python
#Label encoding for gender column
df['gender']=df['gender'].map({'Male':0,'Female':1})
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

scalar = StandardScaler()
minmax=MinMaxScaler()
numCols=['age','annual_income']
df_scalar = scalar.fit_transform(df[numCols])
print(f"scaled data using standard scalar: \n{df_scalar}")
df_minmax = minmax.fit_transform(df[numCols])
print(f"\nscaled data using minmax scalar: \n{df_minmax}")

'''Use Min-Max Scaling when you want values in a fixed range (like 0–1), especially for neural networks or distance-based algorithms without major outliers.
Use StandardScaler when the data is approximately normally distributed or when algorithms assume mean = 0 and standard deviation = 1 (e.g., SVM, Logistic Regression, PCA). '''
```
