#### Approach:
1. Find how much is the missing value % - If the missing % is < 5% then we can drop the records. If the missing % is between 5 and 30, then we need to apply null filling approaches(ffill, bfill, mean etc). If the % is >4= 30 then, drop the entire column.
2. This is not like a mantra or the pre-defines rules for missing values, but this is the recommended approach.
3. It is always recommended to not input values into the target column (selling price). Since our % is < 5%, then we can drop those rows.
4. According to the domain knowledge, in given particular price range, the engine will be of same model. So, we can replace the null values with mode(most repeating one).

### Before we start imputing the values in the columns, we need understand these scenarios:
MCAR - Missing Completely At Random - Just the values are missing on random. Like there is no pattern for this missing. In such scenarios, we can replace the empty values with mean, median or mode.

MAR - Missing At Random - This is like, values in a column can be missing due to some other column. Likea feature "Sunroof". Here the values for sunroof column(example) are missing where the manufacture year < 200. Such a feature is missing in these cars, so null values.

MNAR - Missing Not At Random - This is like, people deliberately miss to include this info. Eg: Accident info or challans info regarding the car. This will impact selling price. This is the hardest scenario to handle, because we need a lot of domain knowledge to predict and fill this.

### Handling Duplicates
Is it better to keep the duplicates or Remove it???? It is better to remove the duplicates, because, when we train the model with duplicates we do:
1. Make the model biased, because we rae traing with same data multiple times.
2. Because, the model get biased, then the metrics produced by model will be inflated.
3. We will also be wasting the memory and computation power of the model.

```python
print(f"Total no.of duplicates present in dataset: {df.duplicated().sum()}")
df=df.drop_duplicates()
print(f"Total no.of duplicates present in dataset after dropping duplicates: Just for checking: {df.duplicated().sum()}")
df.describe()
```
