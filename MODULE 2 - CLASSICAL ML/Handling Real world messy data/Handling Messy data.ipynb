Now I will calculate the "Mean Absolute Error.(MAE)"(Mean/Avg of the difference btw predicted value by model and test data's target data). Now our end goal b4 releasing the model is to have minimum MAE.




### DATA LEAKAGE
During testing time, model will give very good results. The MAE will be low. But when we deploy it in prod, it will break. This will happen when we unkowingly, expose the test data to the model in the training period. So the model will obviously predict the target during testing time. This will lead to less MAE.



### Validation Split
We split the trainging data(80%) into again 80% and 20%. I will use the 2nd 80% for training, 2nd 20% for validating the mode. Once my validation is fair enough, I will use the last 1st 20% will be used for testing. It is like havinhg internal exams before final exam.

### Bias

When you have variation in target, but if that variation is not balanced(Eg: Car's selling price: 100000,110000,120000.....2000000,3000000). Here in example, we have some patthers in the first few numbers, but when it comes to last the numbers are drastic and are not in sync. This is called imbalace in variation, in the target variable. Now the model will try tio find patterns for those entries, where the variation is balaced, leaving it to be biased. Now, if my dataset has 2 sets of data distribusion of seller price. Now when we split the dataset into 80 20, unfortunately ehat if the second set of data goes into testing? Then model will not be able to predict anything testing phase.

Solution for the above problem::: We will divide our training data set into multiple chunks(eg: 5), in 1st iteration, i will use 5th part for testing and 1-4 for training. I will calculate the MAE. 2nd iteration - 1,2,3,5 - training, 4 - testing. Like this i will train my model with all the parts, and test with all the parts. This way, my model would have been trained with all types of data in mhy dataset. After all these scenarios, my Average MAE will deciede my model's performance.


#### how to calculate Mean Absolute Error
***MAE = sum(abs(target values - predicted values))/Total no.of records.***

Benchmark for MAE = First we need to find the Baseline. baseline MAE :
predicted value for every record = avg(target value)
Now MAE = sum(abs(target values - predicted values))/Total no.of records.(Here every target value is avg of target values in the dataset.)
So, if this MAE we found just now is called ***Baseline***.
On applying some ML model for my dataset and if the resultant MAE is > Baseline MAE, then it is the worst model.
