```python
import numpy as np
import pandas as pd
```

```python
data = [170, 165, 180, 172, 168, 160, 175, 171]
```

```python
def calculate_mean(data):
    return sum(data)/len(data)

def calculate_median(data):
    data_sorted = sorted(data)
    n = len(data_sorted)
    m = n // 2

    if n % 2 == 0:
        return (data_sorted[m-1] + data_sorted[m])/2
    else:
        return data_sorted[m]
    
def calculate_std(data):
        mean = calculate_mean(data)
        nominator = sum( (i-mean) ** 2 for i in data)
        variance = nominator / len(data)

        return round(variance ** 0.5,2)


print(f"Mean: {calculate_mean(data)}")
print(f"Median: {calculate_median(data)}")
print(f"Standard Deviation: {calculate_std(data)}")
```
