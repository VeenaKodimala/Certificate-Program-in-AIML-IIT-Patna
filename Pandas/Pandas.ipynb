### SERIES
Series is a datatype of Pandas, it is basically a 1-D array with labels/header(like in excel). We can create custom index to the values of the array.
eg: [1,2,3,4,5] ---> Here usually index of 2 is 1 likein arrays or lists. In series we hcan have a custom index for it like "Veena". This is the biggest advantage of Series.

```python
import pandas as pd
```

```python
pdSeries1 = pd.Series([23.56,78.89,98,100,'Veena'])
print(f"pdSeries1:: {pdSeries1}")
```

```python
##How to create custom Indices##
pdSeriesInd = pd.Series([23.56,78.89,98,100],index=['Rue','Zoe','Ann','May'],name='ScoreRegister')
print(f"pdSeriesInd:: {pdSeriesInd}")

####Creating pandas series from dictionaries####
pdSeriesDict = pd.Series({'Rue':23.56,'Zoe':78.89,'Ann':98,'May':100})
pdSeriesDict
##print(f"Series created from Dictionary:: {pdSeriesDict}")

##How to access the series###
print(f"values:: {pdSeriesDict.values}")
print(f"Indices:: {pdSeriesDict.index}")

print("Ann::: ",pdSeriesDict['Ann'])

# Basic math
print(pdSeriesDict.mean()) #84.17
print(pdSeriesDict.max()) #91.0
print(pdSeriesDict.min()) #73.0
print(pdSeriesDict.sum()) #252.5
#Apply math to all values at once (vectorized no loops needed!)
print(F"pdSeriesDict + 5::: {pdSeriesDict + 5}")
#Adds 5 to every score
print(F"pdSeriesDict * 2::  {pdSeriesDict * 2}")
#Doubles every score
print(F"pdSeriesDict > 80:: {pdSeriesDict > 80}")
#Returns True/False for each value
```

## Lock(location) and Ilock(int-location)

```python
#######SYNTAX FOR ILOC ----- df.iloc[index_start:index_End , column_index_start:column_index_End] -------###########

#Integer location(trick to remember - index loaction)
print(df.iloc[0])

#slicing in iloc
print(df.iloc[0:3])

#slicing in iloc
print(df.iloc[0:3, 0:3])

#slicing in iloc, getting delhi's 1st value
print("Delhi's first value : ",df.iloc[1, 0])
```

```python
##############SYNTAX FOR LOC ------------df.iloc[index_start:index_End/index_names_list , column_names_list] --------------
data = [{"Name": "Alice","Age": 20, "Score": 88.5},
{"Name": "Bob","Age": 21, "Score": 73.0},
 {"Name": "Charlie", "Age": 19, "scofer1": 91.0}]

#NaN Not a Number.
df = pd. DataFrame(data,index=['entry1','entry2','entry3'])
print(df)

print(df.loc[0]) #This will not work because there is no row index called 0.
```

```python
print(f"Using iloc:::: {df.iloc[0:2,[0,2]]}")
#list of row indices and list of column indices
print(f"Using loc:::: {df.loc[['entry1','entry2'],['Name','Score']]}")
```
